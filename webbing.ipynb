```python
with h5py.File('nexus_outputs.h5', 'r') as f:
    norm_dens = f['z_0.0']['density'][:]
    voids = f['z_0.0']['voids'][:]
    filaments = f['z_0.0']['filaments'][:]
    nodes = f['z_0.0']['nodes'][:]
    walls = f['z_0.0']['walls'][:]
    print(f"Suessfully Loaded Stuff")
```

```python
norm_dens_k = fftn(norm_dens)
power_spec = np.abs(norm_dens_k)
```

```python
res = 128
L = 35.0 #cMpc/h
dx = L / res

k_1d = 2 * np.pi * fftfreq(res, d=dx)

kx, ky, kz = np.meshgrid(k_1d, k_1d, k_1d, indexing='ij')
k_squared = kx**2 + ky**2 + kz**2
k_squared[0, 0, 0] = 1.0
phi_k = -norm_dens_k / k_squared
phi_k[0, 0, 0] = 0.0 + 0.0j

pot = np.real(ifftn(phi_k))
```

```python
plt.imshow(pot[0,:,:],origin = "lower", extent = [0,35,0,35])
plt.colorbar(label = r"(cMpc/h)$^2$")
plt.contour(pot[0,:,:], extent=[0, 35, 0, 35], colors='white', alpha=0.2, levels=20)
plt.xlabel("cMpc/h")
plt.ylabel("cMpc/h");
```
